```python
import folium
import geopandas as gpd
import pandas as pd
```

```python
map = folium.Map(location=[23.947429, -27.605951], zoom_start=3)
folium.TileLayer('cartodbpositron').add_to(map)
```

```python
map
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
countries = ["Colombia", "Peru", 
                    "Sierra Leone", 
                    #"Ghana", 
                    "Uganda", 
                    "Kenya","Nigeria","Niger",
                    "Chad",
                    "Jordan",
                    "Lebanon", 
                    #"Iraq",
                    "Syria","Myanmar","Bangladesh", 
                    #"India",
                    "S. Sudan",
                     "Dem. Rep. Congo"
                    ]
map_countries = world[world["name"].isin(countries)]
```

```python
geo_json = folium.GeoJson(map_countries,
                          style_function = lambda feat: {'color':'#57068c', 'fillColor':'#57068c',"weight":1},
                          highlight_function= lambda feat: {'fillColor': 'purple'})
                          #popup=folium.GeoJsonPopup(fields=['name'],labels=False ))

geo_json.add_to(map)
```

```python
map
```

```python
active_countries = ["Colombia", "Peru", 
                    "Sierra Leone", 
                    #"Ghana", 
                    "Uganda", 
                    "Kenya","Nigeria","Niger",
                    "Chad",
                    "Jordan",
                    "Lebanon", 
                    #"Iraq",
                    "Syria","Myanmar","Bangladesh", 
                    #"India",
                    "S. Sudan",
                   "Congo"]
html = []
for i in active_countries: 
    link = "html/"+i+".html"
    html.append(link)
```

```python
country_loc = [
    [4.606880,-74.071838], 
    [-12.046373,-77.042755], 
    [8.4657, -13.2317],
    #[5.603717,-0.186964],Ghana
    [0.347596,32.582520], #Uganda
    [-1.292066,36.821945],
    [9.0765, 7.3986],
    [13.5116,2.1254],
    [12.1348, 15.0557], #Chad
    #[39.9334, 32.8597],Turkey
    [31.9539, 35.9106],
    [33.8938, 35.5018],
    #[31.898043,35.204269],Palestine
    #[33.3152, 44.3661], Iraq
    [33.5138, 36.2765],
    [19.7633, 96.0785],
    [23.8103, 90.4125],
    #[28.6139, 77.2090],India
    #[13.6929, -89.2182] El Salvador
    [4.8594, 31.5713],#"S. Sudan",
    [-4.322447, 15.3105]#"Dem. Rep. Congo", 
    #"Chile", 
    #"Afghanistan", 
    #"Pakistan"
    ]
```

```python
country_loc = pd.DataFrame(country_loc)
df = pd.DataFrame()
df = df.assign(Name = active_countries, html_file = html, Latitude = country_loc[0], Longitude=country_loc[1])
```

```python
for i in range(0,len(df)):
    html="""
    <iframe src=\"""" + df['html_file'][i] + """\" width="850" height="400"  frameborder="0">    
    """
    
    popup = folium.Popup(folium.Html(html, script=True))
    #pushpin = folium.features.CustomIcon('./images/purple_pin.png', icon_size=(15,30))
    folium.Marker([df['Latitude'].iloc[i],df['Longitude'].iloc[i]],
                  popup=popup,icon=folium.Icon(color='darkpurple', icon='info', prefix='fa')).add_to(map)
    
```
