```python
import folium
import geopandas as gpd
import pandas as pd 
```

```python
map = folium.Map(location=[33.06,39.49], zoom_start=3)
```

```python
map
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
world
```

```python
countries = ["Colombia", "Peru", "Siera Leone", "Ghana", "Uganda", "Kenya","Nigeria","Niger","Chad","Turkey","Jordan","Lebanon","Palestine", "Iraq","Syria","Myanmar","Bangladesh","India", "El Savador"]
```

```python
map_countries = world[world["name"].isin(countries)]
```

```python
map_countries
```

```python
geo_json = folium.GeoJson(map_countries,
                         highlight_function= lambda feat: {"fillColor":'purple'})
geo_json.add_to(map)
```

```python
map
```

```python
map.save("TIES Map.html")
```
